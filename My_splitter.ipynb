```python
def inputter(func):
    
    def reader(*args, **kwargs):
        
        txt = input("Type the text or file's path >> ")
        
        #tries to pass the file path, otherwise passes the input string
        try:
            with open(txt, 'r') as t:
                txt = t.read()
                
            return func(txt, *args, **kwargs) 
        
        except FileNotFoundError:
            
            return func(txt, *args, **kwargs)
   
    return reader
```

```python
@inputter
def splitter(txt):

    from string import whitespace as whs, punctuation as pct 

    L = [] 
    word = ''
 
    for c in txt:
        if c not in (whs + pct):
            word += c

        else:
            if c in pct:
                if word:
                    L.append(word)
                word = c
                L.append(word)
                word = ''
                continue

            elif not word:
                continue

            else:
                L.append(word)
                word = ''
    
    return L
```
