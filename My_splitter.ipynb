```python
def inputter(func):
    
    def reader(*args, **kwargs):
        
        testo = input("Type the text or file's path: ")
        
        #tries to pass the file path, otherwise passes the input string
        try:
            with open(testo, 'r') as t:
                testo = t.read()
                
            return func(testo, *args, **kwargs) 
        
        except FileNotFoundError:
            
            return func(testo, *args, **kwargs)
   
    return reader
```

```python
@inputter
def splitter(testo):

    from string import whitespace as whs, punctuation as pct 

    L = [] 
    word = ''
    ex = whs + pct

    for c in testo:
        if c not in ex:
            word += c
        else:
            if not word:
                continue
            else:
                L.append(word)
                word = ''
    if word:        
        L.append(word)
    
    return L
```
